```python
dados_mongo = list(colecao.find({
    'valor' : {'$gte' : 90, '$lte' : 200}
}))

display(dados_mongo)
```

```python
dados_cidade = list(colecao.find({
    'cidade' : {'$in' : ['Ourinhos', 'Assis', 'Pompeia', 'Marilia']}
}))
display(dados_cidade)
```

```python
df = pd.DataFrame(dados_cidade)
df
```

```python
df.isnull().sum()
```

```python
total = df['valor'].sum()
print(total)
```

```python
faturamento_cidade = df.groupby('cidade')['valor'].sum().reset_index()
print(faturamento_cidade)
```

```python
servico = df.groupby('servico')['valor'].sum().reset_index()
print(servico)
```

```python
ticket_medio_geral = df['valor'].mean()
print(ticket_medio_geral)
```

```python
ticket_medio_cidade = df.groupby('cidade')['valor'].mean()
print(ticket_medio_cidade)
```

```python
servico_lucrativo = df.groupby('servico')['valor'].sum().sort_values(ascending=False)
print(servico_lucrativo.head(1))
```

```python
cidade_faturamento = df.groupby('cidade')['valor'].sum().sort_values(ascending=False)
print(cidade_faturamento.head(1))
```

```python
top3_atendimentos = df.sort_values(by='valor', ascending=False).head(3)
print(top3_atendimentos)
```

```python
relatorio = {
    "data_geracao": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "faturamento_total": float(df['valor'].sum()),
    "faturamento_por_cidade": df.groupby('cidade')['valor'].sum().to_dict(),
    "faturamento_por_servico": df.groupby('servico')['valor'].sum().to_dict(),
    "ticket_medio_geral": float(df['valor'].mean()),
    "ticket_medio_por_cidade": df.groupby('cidade')['valor'].mean().to_dict(),
    "servico_mais_lucrativo": df.groupby('servico')['valor'].sum().idxmax(),
    "cidade_maior_faturamento": df.groupby('cidade')['valor'].sum().idxmax(),
    "top_3_atendimentos": df.sort_values(by='valor', ascending=False)
                             .head(3)
                             .to_dict(orient='records')
}

print(relatorio)
```
